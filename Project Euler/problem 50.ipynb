# Problem 50
The prime 41, can be written as the sum of six consecutive primes:

41 = 2 + 3 + 5 + 7 + 11 + 13
This is the longest sum of consecutive primes that adds to a prime below one-hundred.

The longest sum of consecutive primes below one-thousand that adds to a prime, contains 21 terms, and is equal to 953.

Which prime, below one-million, can be written as the sum of the most consecutive primes?

```python
import pandas as pd
```

```python
df=pd.read_excel('My_prime_dataset.xlsx')
```

```python
import numpy as np
```

```python
a=np.array(df.values)
b=a.flatten()
```

```python
primeSet=b[b<=1000000]
primeSet.sort()
primeSet
```

```python
s=0
grt=2
step=0
pset=set()
grtLen=1
lenprimeSet= len(primeSet)

while(step<lenprimeSet):
    pset=set()
    s=0
    for y in primeSet[step:]:
        if(s+y>1000000):
            break
        s+=y

        pset.add(y)
        #Now check if sum is prime
        if(s in primeSet):            
            if(len(pset)>grtLen):
                grt=s
                grtLen=len(pset)
                #print("Set: ",pset)
                print("Grt: ",s,"\nlen(Set): ",len(pset),"\n")
        
        
    step+=1

print(grt)
```
