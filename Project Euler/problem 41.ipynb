# Problem 41
We shall say that an n-digit number is pandigital if it makes use of all the digits 1 to n exactly once. For example, 2143 is a 4-digit pandigital and is also prime.

What is the largest n-digit pandigital prime that exists?

# Solution:

```python
import pandas as pd
```

```python
import primePkg as pp
import numpy as np
```

```python
d=dict()

for i in range(1,10):
    l=[]
    if(i not in d):
        for j in range(1,i+1):            
            l.append(str(j))
        d[i]=l        
d
```

# Since this is about primes it's better to traverse along the primeSet instead
# of checking if it's prime for every number.

```python
df=pd.read_excel('My_prime_dataset.xlsx')
```

```python
a=np.array(df.values)
primeSet=a.flatten()
primeSet.sort()
primeSet=primeSet[::-1]
primeSet
```

# Solution:

```python
s=0
for x in primeSet:    
    s=0
    var=list(str(x))
    l=len(var)
    var.sort()
    if(x==2143):
        print(var)
    if(var==d[l]):
        print(x)
```
